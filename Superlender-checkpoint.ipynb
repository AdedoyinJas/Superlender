```python
def employment_status(df):
    if pd.isnull(df['employment_status_clients']):
        
        if df['bank_account_type'] == 'Current':
            return "Contract"
        elif df['year'] >= 1980 and df['year'] < 1987:
            return "Permanent"
        elif df['year'] >= 1987 and df['year'] < 1991:
            return 'Unemployed'
        elif df['year'] >= 1991 : 
            return 'Student'
        elif df['year'] >= 1974 and df['year'] < 1980:
            return 'Self - employed'
        
    else:
        return df["employment_status_clients"]
```
